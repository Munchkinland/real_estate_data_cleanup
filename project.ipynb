#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# List of populations in the "south belt" of Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a subset of the original DataFrame based on the populations
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)]

# Create a list to store the normalized prices for each population
normalized_prices = []

# Normalize prices for each population and store them in the list
for population in south_belt_populations:
    population_data = south_belt_subset[south_belt_subset['level5'] == population]
    prices = population_data['price']
    # Normalize prices using z-score (standardization)
    normalized_prices.append((prices - prices.mean()) / prices.std())

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Plot histograms for each population with normalized prices
n, bins, patches = ax.hist(
    normalized_prices,
    bins=20,  # You can adjust the number of bins as needed
    density=True,
    alpha=0.7,
    label=south_belt_populations
)

# Customize the plot
ax.set_title("Histograma de precios normalizados")
ax.set_xlabel("Precio Normalizado")
ax.set_ylabel("Frecuencia")
ax.grid(True)

# Add a legend to differentiate populations
ax.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
from ipyleaflet import Map, CircleMarker, basemaps

# Cargar el conjunto de datos
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Convertir las columnas 'latitude' y 'longitude' a números y manejar valores faltantes
ds['latitude'] = ds['latitude'].str.replace(',', '.').astype(float)
ds['longitude'] = ds['longitude'].str.replace(',', '.').astype(float)

# Filtrar y limitar las propiedades en la franja sur de Madrid (ajusta las coordenadas según tus criterios)
# Aquí limitamos las propiedades a las primeras 1000 filas
south_belt_properties = ds[(ds['latitude'] >= 40) & (ds['longitude'] >= -4)].head(1000)

# Crear un mapa centrado en Madrid con un zoom inicial más cercano
madrid_map = Map(center=(40.416775, -3.703790), zoom=7, basemap=basemaps.Stamen.Terrain)

# Define una función para asignar colores a las poblaciones
def get_color(population):
    if population == "Fuenlabrada":
        return "green"
    elif population == "Leganés":
        return "blue"
    elif population == "Getafe":
        return "red"
    elif population == "Alcorcón":
        return "purple"
    else:
        return "gray"  # Otros en gris

# Iterar a través de las propiedades de la franja sur
for index, row in south_belt_properties.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    population = row['level5']  # Utiliza el nombre real de la columna de población
    
    # Obtener el color para la población actual
    color = get_color(population)
    
    # Crear un marcador circular en la ubicación actual
    marker = CircleMarker(location=(lat, lon), radius=5, color=color, fill_color=color, fill_opacity=0.7)
    
    # Agregar el marcador al mapa
    madrid_map.add_layer(marker)

# Mostrar el mapa
madrid_map
```
